```python
import yfinance as yf
import pandas as pd
import time
import logging
```

```python
logging.basicConfig(
    filename="data_fetch.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)
```

```python
TICKERS = [
    "HDFCBANK.NS",   # Banking
    "TCS.NS",        # IT
    "HINDUNILVR.NS", # FMCG
    "RELIANCE.NS",   # Energy
    "MARUTI.NS",     # Automobile
    "SUNPHARMA.NS",  # Pharma
    "TATASTEEL.NS",  # Metals
    "BHARTIARTL.NS", # Telecom
    "LT.NS",         # Infrastructure
    "DMART.NS",      # Retail
    "^NSEI"          # Indexcv
]
```

```python
START_DATE = "2018-01-01"
END_DATE = "2023-12-31"
```

```python
def fetch_with_retry(ticker, retries=3, delay=3):
    for attempt in range(retries):
        try:
            data = yf.download(
                ticker,
                start=START_DATE,
                end=END_DATE,
                progress=False,
                auto_adjust=True
            )

            if data.empty:
                raise ValueError("Empty data returned")

            logging.info(f"Success: {ticker}")
            return data

        except Exception as e:
            logging.warning(f"Attempt {attempt+1} failed for {ticker}: {e}")
            time.sleep(delay)

    logging.error(f"Failed to fetch data for {ticker}")
    return None
```

```python
def validate_data(df, ticker):
    if df is None:
        return False

    if df.isnull().sum().sum() > 0:
        logging.warning(f"Missing values found in {ticker}")

    return True
```

```python
import os
os.makedirs("data", exist_ok=True)

def save_data(df, ticker):
    file_path = f"data/{ticker}.csv"
    df.to_csv(file_path)
    logging.info(f"Saved: {file_path}")
```

```python
all_data = {}
failed_tickers = []

for ticker in TICKERS:
    print(f"Fetching {ticker}...")

    df = fetch_with_retry(ticker)

    if df is not None:

        # Fix multi-index columns
        if isinstance(df.columns, pd.MultiIndex):
            df.columns = df.columns.get_level_values(0)

        # Ensure Close exists
        if 'Close' not in df.columns:
            print(f"Skipping {ticker} - No Close column")
            failed_tickers.append(ticker)
            continue

        if validate_data(df, ticker):
            save_data(df, ticker)
            all_data[ticker] = df
        else:
            failed_tickers.append(ticker)

    else:
        failed_tickers.append(ticker)

    time.sleep(2)

print(" Data collection completed")
print("Successful:", len(all_data))
print("Failed:", failed_tickers)
```

```python
import pandas as pd

price_df = pd.DataFrame()

for ticker, df in all_data.items():
    price_df[ticker] = df['Close']

# Remove missing values
price_df = price_df.dropna()

print(price_df.head())
print(price_df.shape)
```

```python
import numpy as np

# Convert price DataFrame to NumPy array
prices = price_df.values

# Compute log returns
log_returns = np.log(prices[1:] / prices[:-1])

# Convert back to DataFrame (for readability)
log_returns_df = pd.DataFrame(
    log_returns,
    columns=price_df.columns,
    index=price_df.index[1:]
)

# Check output
print(log_returns_df.head())
print(log_returns_df.shape)
```

```python
log_returns_df.isnull().sum()
```

```python
# Mean returns (average daily return of each stock)
mean_returns = np.mean(log_returns, axis=0)

# Covariance matrix (how stocks move together)
cov_matrix = np.cov(log_returns.T)

print("Mean Returns:\n", mean_returns)
print("\nCovariance Matrix shape:", cov_matrix.shape)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1 — Calculate Correlation Matrix
corr_matrix = pd.DataFrame(
    log_returns,
    columns=price_df.columns
).corr()

print("Correlation Matrix:")
print(corr_matrix.round(2))

# Step 2 — Plot Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(
    corr_matrix,
    annot=True,           # show numbers
    fmt='.2f',            # 2 decimal places
    cmap='RdYlGn',        # Red=negative, Yellow=neutral, Green=positive
    center=0,             # center color at 0
    vmin=-1, vmax=1,      # range -1 to +1
    linewidths=0.5,
    square=True
)
plt.title('Stock Correlation Matrix', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('correlation_matrix.png', dpi=150)
plt.show()
print(" Correlation heatmap saved!")

# Step 3 — Export for Tableau
corr_long = corr_matrix.reset_index()
corr_long = corr_long.melt(
    id_vars='index',
    var_name='Stock_2',
    value_name='Correlation'
)
corr_long.columns = ['Stock_1', 'Stock_2', 'Correlation']
corr_long.to_csv('correlation_matrix.csv', index=False)
print(" correlation_matrix.csv exported!")
```

```python
# Simulation settings
num_simulations = 10000
num_days = 252  # trading days in a year

# Equal weights (for now)
num_assets = len(price_df.columns)
weights = np.ones(num_assets) / num_assets

# Initial investment
initial_value = 100000  # ₹1 lakh

# Store final portfolio values
final_values = []

for i in range(num_simulations):

    # Generate random returns (multivariate normal)
    simulated_returns = np.random.multivariate_normal(
        mean_returns,
        cov_matrix,
        num_days
    )

    # Portfolio daily returns
    portfolio_returns = simulated_returns @ weights

    # Convert to cumulative growth
    portfolio_growth = np.exp(np.cumsum(portfolio_returns))

    # Final portfolio value
    final_value = initial_value * portfolio_growth[-1]

    final_values.append(final_value)

# Convert to NumPy array
final_values = np.array(final_values)

print("Simulation complete ")
```

```python
from scipy.stats import skew, kurtosis

# --- Simulated portfolio returns ---
simulated_returns = (final_values - initial_value) / initial_value

sim_skew = skew(simulated_returns)
sim_kurt = kurtosis(simulated_returns)

print("Simulated Skewness:", sim_skew)
print("Simulated Kurtosis:", sim_kurt)


# --- Historical portfolio returns ---
# Use same weights
historical_portfolio_returns = log_returns @ weights

hist_skew = skew(historical_portfolio_returns)
hist_kurt = kurtosis(historical_portfolio_returns)

print("\nHistorical Skewness:", hist_skew)
print("Historical Kurtosis:", hist_kurt)
```

```python
from datetime import datetime, timedelta
import yfinance as yf
import pandas as pd
import numpy as np

def refresh_market_data():
    print(f" Refreshing data at {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    
    TICKERS = [
        'HDFCBANK.NS', 'TCS.NS', 'HINDUNILVR.NS', 'RELIANCE.NS',
        'MARUTI.NS', 'SUNPHARMA.NS', 'TATASTEEL.NS', 'BHARTIARTL.NS',
        'LT.NS', 'DMART.NS', '^NSEI'
    ]
    
    # Always fetch latest 6 years
    end_date = datetime.today().strftime('%Y-%m-%d')
    start_date = (datetime.today() - timedelta(days=6*365)).strftime('%Y-%m-%d')
    
    print(f" Fetching from {start_date} to {end_date}")
    
    price_df = pd.DataFrame()
    failed = []
    
    for ticker in TICKERS:
        try:
            df = yf.download(
                ticker,
                start=start_date,
                end=end_date,
                auto_adjust=True,
                progress=False
            )
            if df.empty:
                failed.append(ticker)
                continue
            price_df[ticker] = df['Close']
            print(f" {ticker} fetched")
        except Exception as e:
            failed.append(ticker)
            print(f" {ticker} failed: {e}")
    
    price_df = price_df.dropna()
    print(f"\n {len(TICKERS)-len(failed)} stocks fetched successfully!")
    if failed:
        print(f"Failed: {failed}")
    
    return price_df
```

```python
def recalculate_all(price_df):
    print("\n📊 Recalculating all metrics...")
    
    # Log returns
    prices = price_df.values
    log_returns = np.log(prices[1:] / prices[:-1])
    mean_returns = log_returns.mean(axis=0)
    cov_matrix = np.cov(log_returns.T)
    weights = np.array([1/log_returns.shape[1]] * log_returns.shape[1])
    
    # Monte Carlo
    num_simulations = 10000
    num_days = 252
    initial_value = 100000
    final_values = []
    
    for i in range(num_simulations):
        sim_returns = np.random.multivariate_normal(mean_returns, cov_matrix, num_days)
        port_returns = sim_returns @ weights
        port_growth = np.exp(np.cumsum(port_returns))
        final_values.append(initial_value * port_growth[-1])
    
    final_values = np.array(final_values)
    print(" Monte Carlo complete!")
    
    # Export simulation CSV
    pd.DataFrame({
        'simulation_id': range(1, num_simulations + 1),
        'final_value': final_values,
        'return_pct': ((final_values - initial_value) / initial_value) * 100
    }).to_csv('monte_carlo_simulations.csv', index=False)
    
    # Export risk metrics
    pd.DataFrame({
        'metric': ['Initial Investment', 'Mean Final Value', 'Median Final Value',
                   'Best Case', 'Worst Case', 'VaR 5%'],
        'value': [
            initial_value,
            np.mean(final_values),
            np.median(final_values),
            np.max(final_values),
            np.min(final_values),
            np.percentile(final_values, 5)
        ]
    }).to_csv('risk_metrics.csv', index=False)
    
    # Export KPI summary
    peak = np.maximum.accumulate(final_values)
    max_dd = ((final_values - peak) / peak).min() * 100
    var_5 = np.percentile(final_values, 5)
    
    pd.DataFrame({
        'KPI': ['Current VaR (5%)', 'Max Drawdown',
                'Expected Return', 'Initial Investment'],
        'Value': [round(var_5, 2), round(max_dd, 2),
                  round(np.mean(final_values), 2), initial_value],
        'Label': [f"₹{var_5:,.0f}", f"{max_dd:.1f}%",
                  f"₹{np.mean(final_values):,.0f}", "₹1,00,000"]
    }).to_csv('kpi_summary.csv', index=False)

    # Export rolling volatility
    returns_df = pd.DataFrame(
        log_returns,
        columns=price_df.columns,
        index=price_df.index[1:]
    )
    rolling_vol = returns_df.rolling(window=30).std() * np.sqrt(252)
    rolling_vol_long = rolling_vol.reset_index()
    date_col = rolling_vol_long.columns[0]
    rolling_vol_long = rolling_vol_long.melt(
        id_vars=date_col,
        var_name='Ticker',
        value_name='Rolling_Volatility'
    )
    rolling_vol_long.columns = ['Date', 'Ticker', 'Rolling_Volatility']
    rolling_vol_long['Date'] = pd.to_datetime(rolling_vol_long['Date']).dt.strftime('%Y-%m-%d')
    rolling_vol_long.dropna().to_csv('rolling_volatility.csv', index=False)
    
    print(f" All CSVs updated!")
    print(f" New VaR 5%:  ₹{var_5:,.0f}")
    print(f" Max Drawdown: {max_dd:.1f}%")
    print(f" Expected:    ₹{np.mean(final_values):,.0f}")
```

```python
# ▶️ Run this cell to refresh everything with latest market data!
price_df = refresh_market_data()
recalculate_all(price_df)

print("\n All CSVs updated! Go to Tableau → Data → Refresh All Extracts")
print(f" Last refreshed: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
```
